```
import keras_ocr
import os
import pandas as pd
pipeline = keras_ocr.pipeline.Pipeline()
```

```
image_list= []
for imagex in os.listdir('/content/Data Files/Dataset/'): 
  path = '/content/Data Files/Dataset/'+ imagex
  image_list.append(path)

print(image_list)
```

```
imagex_list=[]
text_list=[]
for z in image_list:
  images = [keras_ocr.tools.read(url) for url in [z]]
  prediction_groups = pipeline.recognize(images)
  for i in range(len(images)):
    text=[]
    for j in range(len(prediction_groups[i])):
      text.append(prediction_groups[i][j][0])
    text = ' '.join(text)
    text_list.append(text)
    imagex_list.append(z)
  results=pd.DataFrame({"Filename":imagex_list,"Text":text_list})
#results.head(-1)
```

```
results.to_csv('out.csv')
```

```
!pip install vaderSentiment googletrans
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from googletrans import Translator
analyser = SentimentIntensityAnalyzer() # initialize it
translator = Translator()

def sentiment_analyzer_scores(sentence):
   trans = translator.translate(sentence).text # extracting translation text
   score = analyser.polarity_scores(trans) # analyzing the text
   score = score['compound']
   if score >= 0.05:
      return 'Positive'
   elif score > -0.5 and score < 0.05:
      return 'Random'
   else:
      return 'Negative'
   return score

df['sentiment']=df['correct_spell'].apply(lambda x:sentiment_analyzer_scores(x))
#df['sentiment_1']=df['correct_ginger'].apply(lambda x:sentiment_analyzer_scores(x))
```
